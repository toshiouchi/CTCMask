```python
import matplotlib.pyplot as plt

f1 = open( '../model/6-73_train_loss_20240330_214725.csv')
f2 = open( '../model/6-73_val_loss_20240330_214725.csv')

line1 = f1.readline()
line1 = line1.replace( "\n", "")

line2 = f2.readline()
line2 = line2.replace( "\n", "")


i = 0
x = []
loss0 = []
loss1 = []
loss = []
wer = []
v_i = 0
v_x = []
v_loss0 = []
v_loss1 = []
v_loss = []
v_wer = []

while line1:
    #print( line1 )
    line1_split = line1.split(",")
    #print( line1_split)
    i += 1
    x.append( i / 1065 )
    loss0.append( float(line1_split[1].split(' ')[1]) )
    loss1.append( float(line1_split[2].split(' ')[1]) )
    loss.append( float(line1_split[3].split(' ')[1]) )
    wer.append( float(line1_split[4].split(' ')[1]) )
    line1 = f1.readline()
    line1 = line1.replace( "\n", "")
    
while line2:
    #print( line2 )
    line2_split = line2.split(",")
    #print( line2_split)
    v_i += 1
    v_x.append( v_i / 119 )
    v_loss0.append( float(line2_split[1].split(' ')[1]) )
    v_loss1.append( float(line2_split[2].split(' ')[1]) )
    v_loss.append( float(line2_split[3].split(' ')[1]) )
    #print( line2_split[4].split(' ')[2]) 
    v_wer.append( float(line2_split[4].split(' ')[2]) )
    line2 = f2.readline()
    line2 = line2.replace( "\n", "")
   
    
#plt.plot( x, loss)
#plt.title( "Loss")
#plt.ylabel( 'Loss')
#plt.plot( v_x, v_loss)
#plt.xlabel( 'Epochs')
#plt.ylabel( 'Loss')

#plt.show()

plt.plot( x, loss1, label='Train')
plt.plot( v_x, v_loss1, label ='Val' )
plt.title( "Loss1" )
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss1' )
plt.ylim( [0,100])
plt.show()

plt.plot( x, loss0, label='Train')
plt.plot( v_x, v_loss0, label ='Val' )
plt.title( "Loss0" )
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss0' )
plt.ylim( [0,7])
plt.show()

plt.plot( x, loss, label='Train')
plt.plot( v_x, v_loss, label ='Val' )
plt.title( "Loss" )
plt.xlabel( 'Epochs')
plt.ylabel( 'Loss' )
plt.ylim( [0,100])
plt.show()


plt.plot( x, wer, label='Train')
plt.plot( v_x, v_wer, label ='Val' )
plt.title( "WER" )
plt.xlabel( 'Epochs')
plt.ylabel( 'WER' )
plt.ylim( [0,100])
plt.show()

print( v_wer )
```
